## Drop unnecessary columns 

```python
# List of columns to drop
columns_to_drop = ['Notes', 'Status']

# Drop the specified columns
df = df.drop(columns=columns_to_drop)
```

```python
df
```

## Find occurrence of employee names

```python
# Count the occurrences of each unique employee name
employee_name_counts = df['EmployeeName'].value_counts()

# Display the result
print(employee_name_counts)
```

## Find number of unique job titles

```python
# Find the number of unique job titles
unique_job_titles_count = df['JobTitle'].nunique()

# Display the result
print("Number of unique job titles:", unique_job_titles_count)
```

## Find total number of job titles that contain captain

```python
# Filter job titles that contain "captain"
captain_titles = df[df['JobTitle'].str.contains('captain', case=False)]

# Count the total number of captain job titles
total_captain_titles = len(captain_titles)

# Display the result
print("Total number of job titles containing 'captain':", total_captain_titles)
```

```python
df.shape
```

## Display all the employee names from fire department

```python
# Filter the data to select rows with job titles containing "Fire Department"
fire_department_employees = df[df['JobTitle'].str.contains('Fire Department', case=False)]

# Extract and display the employee names from the filtered data
fire_department_employee_names = fire_department_employees['EmployeeName']
print(fire_department_employee_names)
```

## Find the minimum, maximum and average basepay

```python
# Calculate the minimum, maximum, and average "BasePay"
min_basepay = df['BasePay'].min()
max_basepay = df['BasePay'].max()
average_basepay = df['BasePay'].mean()

# Display the results
print("Minimum BasePay:", min_basepay)
print("Maximum BasePay:", max_basepay)
print("Average BasePay:", average_basepay)
```

## Replace "Not provided" in employeename column with NAN

```python
# Replace "Not provided" with NaN in the EmployeeName column
df['EmployeeName'] = df['EmployeeName'].replace("Not provided", np.nan)
```

##  Drop rows having 5missing values

```python
# Drop rows with 5 or more missing values
df = df.dropna(thresh=len(df.columns) - 5)

# Reset the index of the DataFrame after dropping rows
df.reset_index(drop=True, inplace=True)
```

##  Find the job title of ALBERT PARDINI

```python
# Find the job title of "ALBERT PARDINI"
albert_pardini_job_title = df[df['EmployeeName'] == 'ALBERT PARDINI']['JobTitle']

# Display the job title
print("Job title of ALBERT PARDINI:", albert_pardini_job_title.values[0])
```

## How much he makes including benefits

```python
# Find the "TotalPayBenefits" for ALBERT PARDINI
albert_pardini_total_pay_benefits = df[df['EmployeeName'] == 'ALBERT PARDINI']['TotalPayBenefits']

# Display the total pay including benefits
print("Total Pay including benefits for ALBERT PARDINI:", albert_pardini_total_pay_benefits.values[0])
```

## Show the name of the person with highest basepay

```python
# Sort the DataFrame by "BasePay" in descending order
df_sorted_by_basepay = df.sort_values(by='BasePay', ascending=False)

# Extract the name of the person with the highest "BasePay"
person_with_highest_basepay = df_sorted_by_basepay.iloc[0]['EmployeeName']

# Display the name of the person with the highest "BasePay"
print("Name of the person with the highest BasePay:", person_with_highest_basepay)
```

## Find the average basepay of all employees per job title

```python
# Group the data by job title and calculate the mean "BasePay" for each group
average_basepay_per_job_title = df.groupby('JobTitle')['BasePay'].mean()

# Display the result
print(average_basepay_per_job_title)
```

## Find the average basepay of all employees having job title Accountant

```python
# Filter the data for rows with the job title "Accountant"
accountant_data = df[df['JobTitle'] == 'Accountant']

# Calculate the mean "BasePay" for employees with the job title "Accountant"
average_basepay_accountant = accountant_data['BasePay'].mean()

# Display the result
print("Average BasePay for employees with job title 'Accountant':", average_basepay_accountant)
```

## Find the five most common jobs

```python
# Count the occurrences of each job title and select the top five
most_common_jobs = df['JobTitle'].value_counts().head(5)

# Display the result
print("The five most common job titles:")
print(most_common_jobs)
```
