## 1:What is the distribution of the 'Outcome' variable?     

```python
# Plotting the distribution of 'Outcome'
plt.figure(figsize=(8, 6))
df['Outcome'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Outcome')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-Diabetic (0)', 'Diabetic (1)'], rotation=0)
plt.show()
```

## 2:How does the distribution of glucose levels differ between diabetic and non-diabetic individuals?

```python
# Plotting the distribution of glucose levels for diabetic and non-diabetic individuals
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Glucose', hue='Outcome', kde=True, bins=30, palette=['skyblue', 'salmon'])
plt.title('Distribution of Glucose Levels by Outcome')
plt.xlabel('Glucose Levels')
plt.ylabel('Frequency')
plt.legend(title='Outcome', labels=['Non-Diabetic (0)', 'Diabetic (1)'])
plt.show()
```

## 3:Is there a correlation between BMI and the likelihood of diabetes?

```python
# Scatter plot of BMI vs. Outcome (diabetic or non-diabetic)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='Outcome', data=df, hue='Outcome', palette=['skyblue', 'salmon'])
plt.title('BMI vs. Diabetes Outcome')
plt.xlabel('BMI')
plt.ylabel('Outcome')
plt.yticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.legend(title='Outcome', labels=['Non-Diabetic (0)', 'Diabetic (1)'])
plt.show()

# Calculate correlation coefficient
correlation_coefficient = df['BMI'].corr(df['Outcome'])
print(f"Correlation Coefficient between BMI and Outcome: {correlation_coefficient}")
```

## 4:What is the average age of diabetic and non-diabetic individuals?

```python
# Calculate the mean age for diabetic and non-diabetic individuals
mean_age_diabetic = df[df['Outcome'] == 1]['Age'].mean()
mean_age_non_diabetic = df[df['Outcome'] == 0]['Age'].mean()

print(f"Mean age for diabetic individuals: {mean_age_diabetic:.2f} years")
print(f"Mean age for non-diabetic individuals: {mean_age_non_diabetic:.2f} years")
```

## 5:What are the descriptive statistics for each variable in the dataset?

```python
# Display descriptive statistics for each variable
descriptive_stats = df.describe()

# Print the results
print(descriptive_stats)
```

## 6:How does DiabetesPedigreeFunction correlate with the likelihood of diabetes?

```python
# Box plot of DiabetesPedigreeFunction by Outcome
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='DiabetesPedigreeFunction', data=df, palette=['skyblue', 'salmon'])
plt.title('DiabetesPedigreeFunction by Diabetes Outcome')
plt.xlabel('Outcome')
plt.ylabel('DiabetesPedigreeFunction')
plt.xticks([0, 1], ['Non-Diabetic (0)', 'Diabetic (1)'])
plt.show()
```

## 7:How is the age distribution in the dataset? Is there a specific age group more prone to diabetes?

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Plot the age distribution for both diabetic and non-diabetic individuals
sns.histplot(data=df, x='Age', hue='Outcome', kde=True, bins=30, palette=['skyblue', 'salmon'])
plt.title('Age Distribution by Diabetes Outcome')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Outcome', labels=['Non-Diabetic (0)', 'Diabetic (1)'])
plt.show()
```

## 8:Is there an association between age and glucose levels?

```python
# Regression plot of Age vs. Glucose with a linear fit
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Glucose', data=df, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title('Regression Plot of Age vs. Glucose')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.show()
```
