## Which year Has The Highest Average Voting?

```python
# Group the data by 'Year' and calculate the average rating for each year
average_rating_by_year = df.groupby('Year')['Rating'].mean()

# Find the year with the highest average rating
year_with_highest_average_rating = average_rating_by_year.idxmax()
highest_average_rating = average_rating_by_year.max()

print(f"The year with the highest average rating is {year_with_highest_average_rating} with an average rating of {highest_average_rating:.2f}")
```

## In which year there was the Highest Average Revenue?

```python
# Group the data by 'Year' and calculate the average revenue for each year
average_revenue_by_year = df.groupby('Year')['Revenue (Millions)'].mean()

# Find the year with the highest average revenue
year_with_highest_average_revenue = average_revenue_by_year.idxmax()
highest_average_revenue = average_revenue_by_year.max()

print(f"The year with the highest average revenue is {year_with_highest_average_revenue} with an average revenue of ${highest_average_revenue:.2f} million")
```

##  Find The Average Rating For Each Director

```python
# Group the data by 'Director' and calculate the average rating for each director
average_rating_by_director = df.groupby('Director')['Rating'].mean()

# If you want to sort the directors by average rating in descending order:
average_rating_by_director = average_rating_by_director.sort_values(ascending=False)

# Print the result
print(average_rating_by_director)
```

## Display(visualize )Top 10 Lengthy Movies Titles and Runtime

```python
# Sort the DataFrame by 'Runtime (Minutes)' in descending order and select the top 10 rows
top_10_lengthy_movies = df.sort_values(by='Runtime (Minutes)', ascending=False).head(10)

# Extract the movie titles and runtimes
titles = top_10_lengthy_movies['Title']
runtimes = top_10_lengthy_movies['Runtime (Minutes)']

# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
plt.barh(titles, runtimes, color='skyblue')
plt.xlabel('Runtime (Minutes)')
plt.title('Top 10 Lengthy Movies Titles and Runtimes')
plt.gca().invert_yaxis()  # Invert the y-axis to display the longest movie at the top
plt.show()
```

##  Display(visualize )Number Of Movies Per Year

```python
# Group the data by 'Year' and count the number of movies per year
movies_per_year = df['Year'].value_counts().sort_index()

# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
plt.bar(movies_per_year.index, movies_per_year.values, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Per Year')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.show()
```

##  Classify The Movies Based On Ratings(Excellent,Good and Average)

```python
# Define rating thresholds for classification
excellent_threshold = 8.0
good_threshold = 6.0

# Create a new column 'Rating Category' based on the rating thresholds
df['Rating Category'] = pd.cut(
    df['Rating'],
    bins=[0, good_threshold, excellent_threshold, df['Rating'].max()],
    labels=['Average', 'Good', 'Excellent'],
    include_lowest=True
)

# Display the result
print(df[['Title', 'Rating', 'Rating Category']])
```

## Count the Number Of Action Movies

```python
# Filter the DataFrame to include only action movies
action_movies = df[df['Genre'].str.contains('Action', case=False, na=False)]

# Count the number of action movies
number_of_action_movies = len(action_movies)

# Display the result
print(f"The number of action movies in the dataset is: {number_of_action_movies}")
```
