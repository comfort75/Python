## Rename month names from Portuguese to English

```python
# Define a mapping dictionary from Portuguese to English month names
New_Month = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'MarÃ§o': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}



# Replace month names using the mapping dictionary
df['month'] = df['month'].map(New_Month)

# Print the updated DataFrame
print(df)
```

## Which month maximum number of forest fires were reported and visualize them

```python
# Group the data by month and calculate the total number of forest fires for each month
monthly_fires = df.groupby('month')['number'].sum().reset_index()

# Find the month with the maximum number of forest fires
max_month = monthly_fires[monthly_fires['number'] == monthly_fires['number'].max()]

# Print the result
print("Month with the maximum number of forest fires reported:")
print(max_month)

# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
plt.bar(monthly_fires['month'], monthly_fires['number'])
plt.xlabel("Month")
plt.ylabel("Total Number of Forest Fires")
plt.title("Total Number of Forest Fires by Month")
plt.xticks(rotation=45)
plt.show()
```

## The month of JULY has the higest number of forest fire because during this period the ground has reach it capacity to absorb the heat.

## Which Year maximum number of forest fires were reported and visualize them

```python
# Group the data by year and calculate the total number of forest fires for each year
yearly_fires = df.groupby('year')['number'].sum().reset_index()

# Find the year with the maximum number of forest fires
max_year = yearly_fires[yearly_fires['number'] == yearly_fires['number'].max()]

# Print the result
print("Year with the maximum number of forest fires reported:")
print(max_year)

# Create a line chart to visualize the data
plt.figure(figsize=(10, 6))
plt.plot(yearly_fires['year'], yearly_fires['number'], marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Total Number of Forest Fires")
plt.title("Total Number of Forest Fires by Year")
plt.grid(True)
plt.show()
```

## In the past 30years, 2003 has over 42,760,674 cases of forest fire which is highest still date.

## Which states maximum number of forest fires were reported and visualize


```python
# Group the data by state and calculate the total number of forest fires for each state
state_fires = df.groupby('state')['number'].sum().reset_index()

# Find the state with the maximum number of forest fires
max_state = state_fires[state_fires['number'] == state_fires['number'].max()]

# Print the result
print("State with the maximum number of forest fires reported:")
print(max_state)

# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
plt.bar(state_fires['state'], state_fires['number'])
plt.xlabel("State")
plt.ylabel("Total Number of Forest Fires")
plt.title("Total Number of Forest Fires by State")
plt.xticks(rotation=45)
plt.show()
```

## Mato Grosso in Brazil has the highest number of forest fire cases 

## Find the total number of fires reported in the Amazonas


```python
# Filter the data for the Amazonas state
amazonas_data = df[df['state'] == 'Amazonas']

# Calculate the total number of fires reported in Amazonas
total_fires_in_amazonas = amazonas_data['number'].sum()

print("Total number of fires reported in Amazonas:", total_fires_in_amazonas)
```

## Display number of fires reported in Amazonas (year-wise) and visualize 


```python
# Filter the data for the Amazonas state
amazonas_data = df[df['state'] == 'Amazonas']

# Group the Amazonas data by year and calculate the total number of fires reported for each year
amazonas_yearly_fires = amazonas_data.groupby('year')['number'].sum().reset_index()

# Create a line chart to visualize the data
plt.figure(figsize=(10, 6))
plt.plot(amazonas_yearly_fires['year'], amazonas_yearly_fires['number'], marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Total Number of Fires in Amazonas")
plt.title("Total Number of Fires Reported in Amazonas (Year-wise)")
plt.grid(True)
plt.show()
```

## Display number of fires reported in Amazonas (day-wise) and visualize 


```python
# Filter the data for the Amazonas state
amazonas_data = df[df['state'] == 'Amazonas']

# Convert the "date" column to a datetime format
amazonas_data['date'] = pd.to_datetime(amazonas_data['date'])

# Group the Amazonas data by date (day) and calculate the total number of fires reported for each day
amazonas_daily_fires = amazonas_data.groupby(amazonas_data['date'].dt.date)['number'].sum().reset_index()

# Create a bar chart to visualize the data
plt.figure(figsize=(12, 6))
plt.bar(amazonas_daily_fires['date'], amazonas_daily_fires['number'])
plt.xlabel("Date")
plt.ylabel("Total Number of Fires in Amazonas")
plt.title("Total Number of Fires Reported in Amazonas (Day-wise)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Find the total number of fires that were reported in 2015 and visualize based on each month


```python
# Filter the data for the year 2015
data_2015 = df[df['year'] == 2015]

# Group the 2015 data by month and calculate the total number of fires reported for each month
fires_in_2015_by_month = data_2015.groupby('month')['number'].sum().reset_index()

# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
plt.bar(fires_in_2015_by_month['month'], fires_in_2015_by_month['number'])
plt.xlabel("Month")
plt.ylabel("Total Number of Fires in 2015")
plt.title("Total Number of Fires Reported in 2015 (Month-wise)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Average number of fires reported from highest to lowest(state-wise) and visualize



```python
# Group the data by state and calculate the average number of fires reported for each state
state_avg_fires = df.groupby('state')['number'].mean().reset_index()

# Sort the states based on the average number of fires in descending order
state_avg_fires = state_avg_fires.sort_values(by='number', ascending=False)

# Create a bar chart to visualize the data
plt.figure(figsize=(12, 6))
plt.bar(state_avg_fires['state'], state_avg_fires['number'])
plt.xlabel("State")
plt.ylabel("Average Number of Fires Reported")
plt.title("Average Number of Fires Reported (State-wise)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
